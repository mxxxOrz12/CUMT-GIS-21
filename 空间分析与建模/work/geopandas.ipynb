```python
import geopandas as gpd 
from sqlalchemy import create_engine

cq = gpd.read_file('./data/.geojson')

cq.plot()
```

```python
engine = create_engine('postgresql://postgres:1205@localhost:5432/spatialAnalysis')
cq.to_postgis(name="chongqing",con=engine,if_exists='replace')
```

```python
sql = "Select * from chongqing"
chongqing = gpd.read_postgis(sql,con=engine,geom_col='geometry')
chongqing.plot()
```
